```python
import json
import os

import requests
from dotenv import load_dotenv

load_dotenv()

api_key = os.getenv("NYTIMES_API_KEY")
```

```python
category = "technology"
url = 'https://api.nytimes.com/svc/news/v3/content/all/{category}.json'.format(category=category)
params = {'api-key': "{api_key}".format(api_key=api_key), 'limit': 5}
response = requests.get(url,params).text
json_data = json.loads(response)["results"]
```

```python
print(json_data[0])
```

```python
counter = 0
for each_news in json_data:
    counter += 1
    output_msg = (
        f"{counter}. {each_news['title']} - under the section {each_news['section']} "
        f"it was published on {each_news['published_date']}\n"
        f"Summary: {each_news['abstract']}\n"
        f"Read more: {each_news['url']}\n"
    )
    print(output_msg)
```
